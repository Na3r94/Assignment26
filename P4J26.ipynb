```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('sudoku.tif')
plt.imshow(img1, cmap = 'gray')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.equalizeHist(img1)
plt.imshow(img2, cmap = 'gray')
rows, cols = img2.shape
for i in range(rows):
    if img2[i,500] <50:
        a1 = i
        break
for i in range(rows):
    if img2[rows-i-1, 500] <50:
        a2 = i
        break
for j in range(cols):
    if img2[500,j] <50:
        b1 = j
        break
for j in range(cols):
    if img2[cols-j-1, 500] <50:
        b2 = j
        break
        
img3 = img2[a1:rows-a2 , b1:cols-b2]
plt.imshow(img3, cmap= 'gray')
```

```python
images = []
rows, cols = img3.shape
number = 0
for i in range(9):
    for j in range(9):
        img4 = img3[i * 133:(1+i)*133, j* 133:(1+j) * 133]
        rows_img4, cols_img4 = img4.shape
        pix = 0
        for n in range(rows_img4):
            for m in range(cols_img4):
                if img4[n,m] < 50 :
                    pix += 1
                    
                    if pix == 3300:
                        number += 1
                        cv2.imwrite(f'output/sudoku_num/num{number}.jpg', img4)
                        break
```
